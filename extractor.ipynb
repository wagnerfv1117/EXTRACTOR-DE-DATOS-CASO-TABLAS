## EXTRACTOR DE DATOS CASO TABLAS

1. Importar librerías

```python
import pdfplumber
import pandas as pd
from PIL import Image
```

2. leer el archivo en PDF

```python
#Cargar la librería pdfplumber
pdf=pdfplumber.open("input/archivo.pdf")

#cargar pagina _0
p0=pdf.pages[0]
```

3. Extraer la data del archivo PDF

```python
table= p0.extract_table()#extracción de la tabla que se encuentra en  la pagina del PDF

df=pd.DataFrame(table[0:], columns=table[0])#se define el dataframe con la librería pandas en el cual se almacenarán los datos.
for column in ["Municipio","Incidencias"]: # se hace un ciclo for, para  las columnas que detectará el algoritmo, para que muestre toda la información de la tabla.
    df[column] = df[column].str.replace(" ","")# luego de hacer el for muestre toda
```

4. lectura de la información extraída del PDF y guardada en el DataFrame 

```python
df.head(42)# función para que muestre el dataframe completo.
```

5. Exportar la información a un archivo en Excel

```python
df.to_excel('C:/Users/wagne/Downloads/EXTRACTOR DE DATOS CASO TABLAS/output/extraccion_tabla.xlsx')
```
